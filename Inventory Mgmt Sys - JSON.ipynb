```python
# Importing Liabraries
import json,time

# Initialising Variable
sale = ''

# Loading the Inventory Records from JSON to String
fd = open('Record.json', 'r')
js = fd.read()
fd.close()

# Converting String Records to JSON/Dictionaries
record = json.loads(js)

# Displaying Offer Section
print("--------------------------OFFER----------------------------------")
print("1. If your total bill is 250 or above 5% discount on total bill. ")
print("2. If your purchase all quantity of the inventory, 10% discount. ")
print("-----------------------------------------------------------------")
print('')

# Displaying the menu
print("---------------------------MENU----------------------------------")
print("       NOTE : 18% GST WILL BE APPLICABLE TO ALL ITEMS            ")
print("ID: Name                                        | Price | Qn Avl.")
print("-----------------------------------------------------------------")
for key in record.keys():
    print(key, ":", record[key]['Name'], "\t\t|", record[key]['Price'], "\t|", record[key]['Qn'])
print("-----------------------------------------------------------------")
print('')


# Taking User Informations
print('-----------------------User Details------------------------------')
ui_name = str(input("Enter UserName : "))
ui_mail = str(input("Enter Mail ID  : "))
ui_ph   = str(input("Enter Phn No   : "))
print("-----------------------------------------------------------------")
print('')

# Taking User Requirements about the Product Purchase Details
print("------------------------Enter Requirement------------------------")
ui_pr = str(input("Enter Product ID : "))
ui_qn = int(input("Enter Quantity   : "))
print("-----------------------------------------------------------------")
print('')

# If we're having equal or more quantity then user wants
if (record[ui_pr]['Qn'] >= ui_qn):
    
    # Total_unit price
    bill = record[ui_pr]['Price'] * ui_qn
    # gst calculation
    gst1 = bill * 0.18
    # total bill calculation
    total_bill = bill + gst1


    
    # Displaying Purchase Details
    print("-------------------------BILL------------------------------------")
    print("Product Name       : ", record[ui_pr]['Name'])
    print("Price Per Item (₹) : ", record[ui_pr]['Price'], "Rs")
    print("Quantity Requested : ", ui_qn)
    print("-----------------------------------------------------------------")
    print("Unit Price (₹)     : ", bill, "Rs")
    print("GST (18%)          : ", gst1, "Rs")
    
    # checks if total_purchase amount is 250 or more
    if(total_bill >= 250):
        
        # discount calculation
        discount_5 = total_bill*0.05
        # round it to 2 decimal point
        discount_5 = round(discount_5, 2)
        
        print("Discount (5%)      : ", discount_5, "Rs")
        
        # total bill calculation after discount applied
        total_bill = total_bill - discount_5
        
        # round the total bill to 2 decimal points
        total_bill = round(total_bill, 2)
        
        print("Bill (₹)           : ", total_bill, "Rs")
    else:
        print("Bill (₹)           : ", total_bill, "Rs")
    
    print("-----------------------------------------------------------------")
    
    # Generating Sales Structure in CSV
    sale = ui_name+","+ui_mail+","+ui_ph+","+ui_pr+","+record[ui_pr]["Name"]+","+str(ui_qn)+","+str(record[ui_pr]["Price"])+","+str(total_bill)+","+time.ctime() + "\n"

    # Updating Records Dictionary
    record[ui_pr]['Qn'] = record[ui_pr]['Qn'] - ui_qn
    
# If we don't have enough quantity
else:
    
    # Displaying Info about the quantity
    print("Sorry, We are not having enough Quantity in our Inventory.")
    print("We're only having", record[ui_pr]['Qn'],"quantity.")
    print("-----------------------------------------------------------------")
    
    # If user wants to purchase whole Quantity for that product
    ch = str(input('Press Y/y to Purchase: '))
    
    if (ch == 'Y' or ch == 'y' ):
        
        # total unit price calculation
        bill_2 = record[ui_pr]['Price'] * record[ui_pr]['Qn']
        
        # gst calculation on total unit price
        gst2 = bill_2 * 0.18 
        # round total unit price bill to 2 decimal point 
        gst2 = round(gst2, 2)
        
        # total bill calculation
        total_bill_2 = bill_2 + gst2
        
        # disount calculation on total bill
        discount_10 = total_bill_2*0.10
        # round discount to 2 decimal points
        discount_10 = round(discount_10, 2) 
        # final bill calculate
        total_bill_2 = total_bill_2 - discount_10
        
        # Displaying Purchase Details
        print("")
        print("-------------------------BILL------------------------------------")
        print("Product Name       : ", record[ui_pr]['Name'])
        print("Price Per Item (₹) : ", record[ui_pr]['Price'], "Rs")
        print("Quantity Requested : ", record[ui_pr]['Qn'])
        print("-----------------------------------------------------------------")
        print("Unit Price (₹)     : ", bill_2, "Rs")
        print("GST (18%)          : ", gst2, "Rs")
        
        # round final bill to 2 decimal point
        total_bill_2 = round(total_bill_2, 2)
        
        print("Discount (10%)     : ", discount_10, "Rs")
        print("Total Bill (₹)     : ", total_bill_2, "Rs")
        print("-----------------------------------------------------------------")
        print('')
        
        # Generating Sales Structure in CSV
        sale = ui_name+","+ui_mail+","+ui_ph+","+ui_pr+","+record[ui_pr]["Name"]+","+str(total_bill_2)+","+time.ctime() + "\n"

        # Updating Records Dictionary
        record[ui_pr]['Qn'] = 0
     
    # If user pressed anything other than 'Y'/'y'
    else:
        print("Thank You!")

# Converting dictionary to String
js = json.dumps(record)

# Saving Records in JSON file
fd = open('Record.json','w')
fd.write(js)
fd.close()

# Saving Records in Sales File
fd = open('Sales.txt', 'a')
fd.write(sale)
fd.close()

print('')
print("-----------------------------------------------------------------")
print("           Thanks for your order, Inventory Updated!             ")
print("-----------------------------------------------------------------")
```
